# Merging DataFrames
Now there are two ways to merge DataFrames.

1) Cocatanate the two dataframes horizontally

2) Merge them horizontally

Lets understand the codes ahead to understand how does these work

# Vertically Merging Two DataFrame

Now lets discuss about a method here that helps us in this process. The method we are going to use is called 
###### merge()
The general syntax of using it is
###### pandas.merge(< Name of DataFrame1 On Left >,< Name Of DataFrame2 on Right >, how="< VALUE >", left_index="True", right_index="True")
We will soon see what all can we store into the value in the how attribute of the merge method. Other than that we will soon see cases where we will not use the left_index and the right_index attributes. Let now deep dive into code and understand what we are talking about.

```python
#Lets first declare two dataframes we are going to merge
#Dataframe1: List of name of students studying in KIIT University and thier schools in which they are studying in.
import pandas as pd
dataframe1=pd.DataFrame([{"Name":"Ashutosh Kumar Singh","School":"Computer Science"},
           {"Name":"Harshal Ambasta","School":"Mechanical Engineering"},
           {"Name":"Ananya Agarwal","School":"Civil Engineering"}])
#Dataframe2:List of name of staff and students who work as staff in KIIT University with the roles they are working in
dataframe2=pd.DataFrame([{"Name":"Ashutosh Kumar Singh","Role":"ML Lead"},
           {"Name":"Ashish Pothal","Role":"Graphic Designer"},
           {"Name":"Sudhanshu Jha","Role":"Student Coordinator"}])
#Lets now set the indexes as names in both datasets so that we can merge them on thier comonality that we are going to call when
#we will set the left and right indexes true in our merge method.
dataframe1=dataframe1.set_index("Name")
dataframe2=dataframe2.set_index("Name")
#Let just print them
print(dataframe1)
print(dataframe2)
```

```python
#Lets move onto the second step. Lets now merge these dataframe. Question is what should be the criteria of merging. We want the
#name of students and staff in our merged dataframe with names column in both the frames merged into one. If you remember set 
#thoery this feat can be achieved by taking a union of both of the dataframes. 

#So for taking union of the two data frames we set how="outer". Since the base on which the merger is taking place is "Name" 
#which also happens to be our index, we set the left and right index value to be true. Lets see how does this happen.

unionDataframe=pd.merge(dataframe1,dataframe2,how="outer",left_index=True,right_index=True)
print(unionDataframe)
```

What we observe above is that the two dataframes got merged. The names common in both the dataframe occur together while uncommon ones are also there. They can be identified via NaN are thier column values.

```python
#Lets change the question here. Now lets say we only want those students in our new dataframe who also work as staff. In set 
#theory we can do this via intersection.

#So for taking intersection of the two data frames we set how="inner". Since the base on which the merger is taking place is
#"Name" which also happens to be our index, we set the left and right index value to be true. Lets see how does this happen.

intersectionDataFrame=pd.merge(dataframe1,dataframe2,how="inner",left_index=True,right_index=True)
print(intersectionDataFrame)
```

What we observe here is the only student common in both the student and staff dataframes.


Lets now dicuss about two more values we can pass through the how attribut of merge method.
###### left
###### right
To understand how they work we need to use examples that lie ahead of us.

```python
#Now here is another merging case using. First let make it clear that our left dataframe is student and the right dataframe is
#staff. Believe me order matters. Lets say we want get the data of all the staff, but we also want the data of those children
#that work as staff. For that we use the "right" value as staff dataframe is on the right. What it will do is that it will 
#definatly print all the value of right dataframe but will add values of left frame that sit common with right dataframe. Lets
#see an example to understand better.

rightDataframe=pd.merge(dataframe1,dataframe2,how="right",left_index=True,right_index=True)
print(rightDataframe)
```

As observable since Ashutosh Kumar Singh was common in both the dataframe we have both the school and row of him being printed. Also observe rest data here in the dataframe is of the rest staff members of the University

```python
#Now here is another merging case using. First let make it clear that our left dataframe is student and the right dataframe is
#staff. Believe me order matters. Lets say we want get the data of all the students, but we also want the data of those children
#that work as staff. For that we use the "left" value as student dataframe is on the right. What it will do is that it will 
#definatly print all the value of left dataframe but will add values of right frame that sit common with left dataframe. Lets
#see an example to understand better.


leftDataframe=pd.merge(dataframe1,dataframe2,how="left",left_index=True,right_index=True)
print(leftDataframe)
```

As observable since Ashutosh Kumar Singh was common in both the dataframe we have both the school and row of him being printed. Also observe rest data here in the dataframe is of the rest students of the University


Recall that at the start of this notebook, I mentioned how the usage of the left and right index attributes were temporary and very soon we were going to replace them. Lets move on that front. Let get ourselves introduced to the attribute 
###### on
The general syntax of usage is
###### on="Name of the column on basis of which the merger of the two dataframes is going to take place"
Now whats the advantage of using this attribute? Understand that till now whenever we were merging 2 dataframes we were merging them on the base of thier indexes. If we wanted any other column in place of the index one, we used to set that column as index and then operate. What __on__ attribute does is that is removes that constrain of index column and allows us merge two data frames on the base of several column, whichever is passed through that attribute.

Lets see examples.

```python
#Let remove name column from the indexes
dataframe1=dataframe1.reset_index()
dataframe2=dataframe2.reset_index()
#Let solve for example 3 .
dataframeRight=pd.merge(dataframe1,dataframe2,how="right",on="Name")
dataframeRight
```

As observable, our system is working and is giving same output as observed in the left_index and right_index case.


Lets move on and lets dicuss something else. Lets discuss a case where we have conflicting datas stored in the dataframes we are working with. Let see what I mean.

```python
students=pd.DataFrame([{"Name":"Ashutosh","School":"Computer Science","Location":"Campus 15"},
                     {"Name":"Aditya","School":"Information Technology","Location":"Campus 17"},
                     {"Name":"Harshal","School":"Mechanical","Location":"Campus 8"}])
staff=pd.DataFrame([{"Name":"Ashutosh","Role":"ML Lead","Location":"Campus 14"},
                  {"Name":"Harshal","Role":"Aerospace Engineer","Location":"Campus 5"},
                  {"Name":"Samal","Role":"Prof. Maths","Location":"Campus 17"}])
```

Observe above. In above dataframes we have two columns common. Name and Location. But observe one thing in Location columns of the two dataframes. Yes they are different, unlike names. So what will happen if we merge the two dataframes together? Lets see

We will left merge on Name.

```python
merger=pd.merge(students,staff,how="left",on="Name").set_index("Name")
merger
```

What do we observe above. We see that the locations in the left dataframe are given by the Location_x and the locations in the right dataframe are given by Location_y.

That how python works. The merge function common columns with different values by appending an _x or _y to help differentiate between which index went with which column of data. The _x is always the left DataFrame information, and the _y is always the right DataFrame information.


Lets talk about merging dataframes. But this time we will merge the dataframe using two columns, not one like we have been doing it till now. For that, inplace of finding a single column name on __on__ attribute we pass a list of names of columns on which the merger is to be undertaken.

Lets understand it using an example.

```python
students=pd.DataFrame([{"First Name":"Ashutosh","Last Name":"Singh","School":"Computer Science","Location":"Campus 15"},
                     {"First Name":"Aditya","Last Name":"Dey","School":"Information Technology","Location":"Campus 17"},
                     {"First Name":"Harshal","Last Name":"Ambasta","School":"Mechanical","Location":"Campus 8"}])
staff=pd.DataFrame([{"First Name":"Ashutosh","Last Name":"Singh","Role":"ML Lead","Location":"Campus 14"},
                  {"First Name":"Harshal","Last Name":"Ambasta","Role":"Aerospace Engineer","Location":"Campus 5"},
                  {"First Name":"Sapan","Last Name":"Samal","Role":"Prof. Maths","Location":"Campus 17"}])
#Lets just print intersection of the two dataframes.
intersection=pd.merge(students,staff,how="inner",on=["First Name","Last Name"])
intersection
```

As observable, in the above dataframe, we want the merger via two columns namely First Name and Last Name. So we passed both of them upon the __on__ attribute in the form of list.

# Horizontally Merging Two DataFrames

Now lets dicuss about the method that helps us here. The method is called
###### concat
The general syntax of using concat is
###### pd.concat( [ DataFrame1,DataFrame2 ] )
Lets see a few example to understand how does this work.

```python
%%capture
#Lets First Import the datasets we are going to work with.
df_2012 = pd.read_csv("assets/MERGED2012_13_PP.csv", error_bad_lines=False)
df_2013 = pd.read_csv("assets/MERGED2013_14_PP.csv", error_bad_lines=False)
```

```python
#Lets print them.
df_2012.head()
```

```python
df_2013.head()
```

```python
#Lets find the lenght of the two dataframes so that we could verify thier lenghts later after we merge them
lenght2012=len(df_2012)
lenght2013=len(df_2013)
totalLenght=lenght2012+lenght2013
totalLenght
```

```python
#Let concatanate the two datasets
newDataSet=pd.concat([df_2012,df_2013])
newDataSet.head()
```

```python
#Let see thier lenght
len(newDataSet)
```

As observable our two lenght are equal, hence concatanation is successfull

Now that we have merged the two dataframes we are not able to guess that which data in the new dataframe belong to which original dataframe. Well that is something we can solve using the keys attribute in the __concat__ method.

Let see examples to understand what is being said.

```python
pd.concat([df_2012,df_2013],keys=["2012","2013"])
```

As observable we are now able to differentiate between data from 2012 and 2013.

```python
# Now we have the indices as the year so we know what observations are from what year. You should know that
# concatenation also has inner and outer method. If you are concatenating two dataframes that do not have
# identical columns, and choose the outer method, some cells will be NaN. If you choose to do inner, then some
# observations will be dropped due to NaN values. You can think of this as analogous to the left and right
# joins of the merge() function.
```
