Lets give you some lecture and revisionism on previously learned concepts. Lets talk about __map__.
The general syntax is,
###### map(<  Function Name >,< Iterable >)
As observable above, the parameters passed are iterable such as lists, tuples,etc and a function name which generally signifies the function in which the iterable is passed. This function then works on each and every element of the iterable, and gives us iterable itself but a changed one.

In dataframes there is method called __applymap__. It takes dataframe name as one of its parameter and name of function through which dataframe is to be passed as another parameter. The general syntax of usage is
###### applymap(< Function Name >, < Dataframe Name >)
This method operates on each and every cell of dataframe a returns an output.

Another option is the __apply__ method. It provides us the option of operating over rows and columns in place of operating over each and every cell. It alse takes 2 parameter, first being the function name, and second being the axis. The general syntax of usage is,
###### < DataFrame >.apply(< Function >, axis=)
Lets see a few examples and understand.


###### Lets do this, lets find the max population estimates for the years 2010,2011,2012,2013,2014,2015

```python
def max_min(dataFrame):
    selectedColumns=dataFrame[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    return pd.Series({"max":np.max(selectedColumns),"min":np.min(selectedColumns)})
```

```python
popDataSet=pd.read_csv("assets/census.csv")
popDataSet.apply(max_min,axis="columns").head()
```

Well here we returned a series only. Well in place of doing that lets do one thing that we should return a dataframe instead but the max and min columns must added to it.

```python
def max_min_2(dataFrame):
    selectedColumns=dataFrame[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    dataFrame["Max"]=np.max(selectedColumns)
    dataFrame["Min"]=np.min(selectedColumns)
    return(dataFrame)
```

```python
#Lets call the function
popDataSet.apply(max_min_2,axis="columns").head()
```

Lets move a step ahead and discuss about lambas. Again another revisionism here.

The general syntax of using lambas is as follow
###### < Variable >=lamba x: < Any Operation With x >
As examplar expression is of lambda is

__lambda x:x*2__

This expression doubles any number that is being passed through x.

Lambdas primarily create ad-hoc functions without requiring the default/orthodox function declaration methodology. 

We now are going to discuss casses where we are going to use lambda functions for data manipulation along with map.

Lets see a few examples.


I believe that the map and apply functions work the same manner. Why?

Both of these functions take two parameters, one being the function on which futher operations are to be carried and second parameter is an iterable. For previous example, we saw we passed a list as the iterable in the map function.

But for apply we pass axis.

Lets deep dive into axis and understand how do they work.

axis stores either row or columns. axis generally tells the interpreter to how to iterate over a set a data. If we pass columns as axis it will iterate over a row, column wise.

Similarly if we pass row as axis value we will find that we get iterations over rows to rows in a common column.

```python
#Lets do what we did in previous question but this time using lambda functions.
rows = ['POPESTIMATE2010', 'POPESTIMATE2011', 'POPESTIMATE2012', 'POPESTIMATE2013','POPESTIMATE2014', 
        'POPESTIMATE2015']
popDataSet.apply(lambda x:np.max(x[rows]),axis=1).head()
```

###### Let's say we want to divide the states into four categories: Northeast, Midwest, South, and West We can write a customized function that returns the region based on the state the state regions information is obtained from Wikipedia.

```python
def get_state_region(x):
    northeast = ['Connecticut', 'Maine', 'Massachusetts', 'New Hampshire', 
                 'Rhode Island','Vermont','New York','New Jersey','Pennsylvania']
    midwest = ['Illinois','Indiana','Michigan','Ohio','Wisconsin','Iowa',
               'Kansas','Minnesota','Missouri','Nebraska','North Dakota',
               'South Dakota']
    south = ['Delaware','Florida','Georgia','Maryland','North Carolina',
             'South Carolina','Virginia','District of Columbia','West Virginia',
             'Alabama','Kentucky','Mississippi','Tennessee','Arkansas',
             'Louisiana','Oklahoma','Texas']
    west = ['Arizona','Colorado','Idaho','Montana','Nevada','New Mexico','Utah',
            'Wyoming','Alaska','California','Hawaii','Oregon','Washington']
    
    if x in northeast:
        return "Northeast"
    elif x in midwest:
        return "Midwest"
    elif x in south:
        return "South"
    else:
        return "West"
```

```python
# Now we have the customized function, let's say we want to create a new column called Region, which shows the
# state's region, we can use the customized function and the apply function to do so. The customized function
# is supposed to work on the state name column STNAME. So we will set the apply function on the state name
# column and pass the customized function into the apply function
df['state_region'] = df['STNAME'].apply(lambda x: get_state_region(x))
```
