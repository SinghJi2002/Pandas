So we have found the means, we have found the databases of late and early submitters. Lets now start something we actually are here to do in this section. Lets perform __Hypothesis Testing__. Lets run t-test on dataset we have just created.

The first step of hypothesis testing is forming an hypothesis. Lets examine the current data we are working on. We here are comparing how different the sets of assigment1_grades are for 2 subDataFrames we formed. Our null hypothesis is that the population involved in either case is are not different, while alternate hypothesis is that they are different. Lets move on ahead with these hypothesis.

What is t-test? 

Well it refers to tests run to compare means of two datagroups.

The general syntax of usage is,

__sp.ttest_ind(arg1,agr2)__

arg1 and arg2 are the two datasets whose means are to be compared. After the comparison tasks are carried out we get a tuple that has 2 values.

1) __t-statistics__: This element tells us the differnce of the two means of the two databases. The higher it is, the more different and independent the mean of arg1 and arg2 are.

2) __p-value__: p-value simply measures probablity of obtaining such a difference(t-statictic) if null hypothesis were true. The higher is the value the more different the means are and vice versa.

Note that there is a concept of threshold value here. We generally assume the threshold value to be 0.05 or 0.01(1 % or 5%). If the p value obtained is less than this, then we drop the null statement and declare it false or else, we adopt the null statement and declare it true.

Now once we obtain the p-value, we compare it with the threshold value. If p-value>threshold value, we see that null hypothesis get proven true. Else the alternative hypothesis rules.

Hope you understood. Lets now see this in action.

```python
from scipy.stats import ttest_ind
ttestRun=ttest_ind(earlySubmitdat["assignment1_grade"],latesubmitDat["assignment1_grade"])
print("The t-statistics is {} and the p-value is {}".format(ttestRun[0],ttestRun[1]))
```

Since the p value is greater than 0.05 threshold, we see that the null statement is adopted.

```python
#Lets check the same for all assignments
ttestRun=ttest_ind(earlySubmitdat["assignment2_grade"],latesubmitDat["assignment2_grade"])
print("The t-statistics is {} and the p-value is {}".format(ttestRun[0],ttestRun[1]))

ttestRun=ttest_ind(earlySubmitdat["assignment3_grade"],latesubmitDat["assignment3_grade"])
print("The t-statistics is {} and the p-value is {}".format(ttestRun[0],ttestRun[1]))

ttestRun=ttest_ind(earlySubmitdat["assignment4_grade"],latesubmitDat["assignment4_grade"])
print("The t-statistics is {} and the p-value is {}".format(ttestRun[0],ttestRun[1]))

ttestRun=ttest_ind(earlySubmitdat["assignment5_grade"],latesubmitDat["assignment5_grade"])
print("The t-statistics is {} and the p-value is {}".format(ttestRun[0],ttestRun[1]))

ttestRun=ttest_ind(earlySubmitdat["assignment6_grade"],latesubmitDat["assignment6_grade"])
print("The t-statistics is {} and the p-value is {}".format(ttestRun[0],ttestRun[1]))
```

Ok, so it looks like in this data we do not have enough evidence to suggest the populations differ with respect to grade. Let's take a look at those p-values for a moment though, because they are saying things that can inform experimental design down the road. For instance, one of the assignments, assignment 3, has a p-value around 0.1. This means that if we accepted a level of chance similarity of 11% this would have been considered statistically significant. As a research, this would suggest to me that there is something here worth considering following up on. For instance, if we had a small number of participants (we don't) or if there was something unique about this assignment as it relates to our experiment (whatever it was) then there may be followup experiments we could run.


P-values have come under fire recently for being insuficient for telling us enough about the interactions which are happening, and two other techniques, confidence intervalues and bayesian analyses, are being used more regularly. One issue with p-values is that as you run more tests you are likely to get a value which is statistically significant just by chance.

```python
#Let now move toward another example. First lets generate dataframe via a new technique.
dataframe1=pd.DataFrame([np.random.random(100) for x in range(100)])
#The command generate a dataframe of 100rows and 100columns, with each value stored in them is between 0 and 1
dataframe2=pd.DataFrame([np.random.random(100) for x in range(100)])
```

Lets calculate difference of mean of each and every column of the two dataframes. We have our null hypothesis as there is difference between the corresponding columns and alternate hypothesis that there is differnce between the two corresponding columns.

```python
#lets first generate the two dataframes.
dataframe1.head()
```

```python
dataframe2.head()
```

```python
#Let declare a function to do our work
def ttester(alpha,significantCols):
    #Alpha declaration is being conducted in order to specify, the point above which we consider a given data significant.
    for cols in dataframe1.columns:
        #Since columns of both the elements are the same, we iterate through the columns of one of them.
        tval,pval=ttest_ind(dataframe1[cols],dataframe2[cols])
        if pval>alpha:
            print("The column {} is considered proving null hypthesis with p-value {}".format(cols+1,pval))
            significantCols=significantCols+1
    print("There are {} null hypothetically true cols making their percentage to be {}".format(significantCols,(significantCols/len(dataframe1.columns))*100))
#Lets call the function above.
ttester(0.1,0)
```

Since there are 94 percent columns showing no difference only 6 percent of columns are different to one another. Hence only 6% columns are statistically significant.

So, keep this in mind when you are doing statistical tests like the t-test which has a p-value. Understand that this p-value isn't magic, that it's a threshold for you when reporting results and trying to answer your hypothesis. What's a reasonable threshold? Depends on your question, and you need to engage domain experts to better understand what they would consider significant.
