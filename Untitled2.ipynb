```python
import pandas as pd
import numpy as np
```

```python
def answer_one():
    #For opening an excel file we use the following syntax.
    Energy=pd.read_excel("assets/Energy Indicators.xls")
    Energy=Energy[18:245]
    #Now abaove code is a step to actually remove the header and footer of excel sheets. See into real excel file that main data
    #begins at 18 and ends at row 245.
    
    #Lets now delete unwanted columns
    Energy=Energy.drop(Energy.columns[0],axis=1)
    Energy=Energy.drop(Energy.columns[0],axis=1)
    #U might be seeing a completely different approach here. Well Note that we use this when we delete columns by index rather 
    #than thier names.
    
    #Let now rename our columns
    Energy=Energy.rename(columns={Energy.columns[0]:"Country",Energy.columns[1]:"Energy Supply",Energy.columns[2]:"Energy Supply per Capita",Energy.columns[3]:"% Renewable"})
    
    #Lets convert all the Energy Supply Available in Petajoules to Gigajoules
    Energy["Energy Supply"]=Energy["Energy Supply"]*1000000
    
    #Lets replace ... with np.NaN
    Energy=Energy.replace('...',np.NaN)
    
    #Renameing rows.
    for i in range(18,245,1): 
        if Energy["Country"][i]=="Republic of Korea":
            Energy["Country"][i]="South Korea"
        elif(Energy["Country"][i]=="United States of America20"):
             Energy["Country"][i]="United States"
        elif(Energy["Country"][i]=="United Kingdom of Great Britain and Northern Ireland19"):
             Energy["Country"][i]="United Kingdom"
        elif(Energy["Country"][i]=="China, Hong Kong Special Administrative Region3"):
             Energy["Country"][i]="Hong Kong"
        elif(Energy["Country"][i]=="Japan10"):
            Energy["Country"][i]="Japan"
        elif(Energy["Country"][i]=="China2"):
            Energy["Country"][i]="China"
        elif(Energy["Country"][i]=="France6"):
            Energy["Country"][i]="France"
        elif(Energy["Country"][i]=="Italy9"):
            Energy["Country"][i]="Italy"
        elif(Energy["Country"][i]=="Spain16"):
            Energy["Country"][i]="Spain"
        elif(Energy["Country"][i]=="Australia1"):
            Energy["Country"][i]="Australia"
        elif('(' in str(Energy["Country"][i])):
            Energy["Country"][i]=Energy["Country"][i].split('(')[0].strip()
        else:
            continue
    
    #Importing new datafile GDP
    GDP=pd.read_csv("assets/world_bank.csv")
    
    #lets cut out the important sections of our datafile
    GDP=GDP[4:268]
    
    #lets set the column headers for the dataframe
    GDP.columns=['Country','Country Code','Indicator Name','Indicator Code',1960,1961,1962,1963,1964,1965,1966,1967,1968,1969,1970,1971,1972,1973,1974,1975,1976,1977,1978,1979,1980,1981,1982,1983,1984,1985,1986,1987,1988,1989,1990,1991,1992,1993,1994,1995,1996,1997,1998,1999,2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015]
    
    #Lets rename some of the elements under the Country Column
    for i in range(4,268,1):
        if GDP["Country"][i]=="Korea, Rep.":
            GDP["Country"][i]="South Korea"
        elif GDP["Country"][i]=="Iran, Islamic Rep.":
            GDP["Country"][i]="Iran"
        elif GDP["Country"][i]=="China, Hong Kong Special Administrative Region":
            GDP["Country"][i]="Hong Kong"
        else:
            continue
    GDP=GDP[['Country','Country Code','Indicator Name','Indicator Code',2006,2007,2008,2009,2010,2011,2012,2013,2014,2015]]
            
    #Let import another data file
    ScimEn=pd.read_excel("assets/scimagojr-3.xlsx")
    ScimEn=ScimEn[0:15]
    
    #Lets merge the 3 dataframes
    mergedDataFrame=pd.merge(ScimEn,GDP,how="inner",on="Country")
    remergedDat=pd.merge(mergedDataFrame,Energy,how="inner",on="Country").set_index("Country")
    remergedDat=remergedDat.drop(remergedDat.columns[7],axis=1)
    remergedDat=remergedDat.drop(remergedDat.columns[7],axis=1)
    remergedDat=remergedDat.drop(remergedDat.columns[7],axis=1)
    remergedDat=remergedDat.sort_values("Rank",ascending=True)
    return(remergedDat)
    raise NotImplementedError()        
```

```python
def answer_three():
    remergedDat=answer_one()
    #Lets now solve our question.
    avgGDP=np.mean(remergedDat[[2006,2007,2008,2009,2010,2011,2012,2013,2014,2015]],axis=1)
    #In general if we simple would have passed the columns with mean function, we would have got function of each column along
    #the row. But as we set axis=1,we get mean of rows along with the columns.
    return(avgGDP.sort_values(ascending=False))
    raise NotImplementedError()
```

```python
answer_three()
```
